```python
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ******* Set the directory you wish to store the dataset *******
data_path = '/Users/libbystephan/Documents/School/Stat 424'
mnist_transform = transforms.Compose([transforms.ToTensor()])\n
celebA_transform = transforms.Compose([
        transforms.Resize((64, 64)),  # Resize the image to 64x64
        transforms.ToTensor()         # Convert the image to a tensor
    ])

celebA_dataset = datasets.CelebA(root=data_path, download=True, transform=celebA_transform) 
mnist_dataset = datasets.MNIST(root=data_path, train=True, download=True, transform=mnist_transform)
```

```python
# How to Access Images and Attributes from the Datasets

celeb_image, attributes = celebA_dataset[0] # the first observation in this case
mnist_image, num = mnist_dataset[0]         # the first observation in this case

# checking the data types
print(celeb_image) # of type PIL image
print(mnist_image) # tensor of the matrix of black-white (luminescence) values, of shape [1,28,28] ie 28x28 pixels, first dim useless
print(attributes)  # tensor of the attributes (see the list of the corresp. attributes)
print(num)         # of type int
```

```python
# How to Display Images

# for CelebA dataset

# Convert PIL Image to NumPy array
celeb_image_np = np.array(celeb_image)

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(celeb_image_np)  # Matplotlib handles RGB automatically
plt.axis("off")  # Hide axes for a cleaner image
plt.title("CelebA Image")
plt.show()
```

```python
# for MNIST Dataset

# Convert tensor to NumPy for visualization
image_np = mnist_image.squeeze().numpy()  # squeeze removes the useless first dimension

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(image_np, cmap="gray")  # Use grayscale colormap
plt.axis("off")
plt.title(f"MNIST Digit: {num}")
plt.show()
```

```python
# Splitting the Color Channels for CelebA

red = celeb_image_np[:, :, 0]
green = celeb_image_np[:, :, 1]
blue = celeb_image_np[:, :, 2]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Show Red Channel
axes[0].imshow(red, cmap="Reds")
axes[0].set_title("Red Channel")
axes[0].axis("off")

# Show Green Channel
axes[1].imshow(green, cmap="Greens")
axes[1].set_title("Green Channel")
axes[1].axis("off")

# Show Blue Channel
axes[2].imshow(blue, cmap="Blues")
axes[2].set_title("Blue Channel")
axes[2].axis("off")

plt.show()
```
